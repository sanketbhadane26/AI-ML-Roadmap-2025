```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna('Unknown')
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Family size']=df['SibSp']+df['Parch']+1
df['Family size']
```

```python
df['Is child']=df['Age'].apply(lambda x: 1 if x<16 else 0)
df['Is child']
```

```python
df['Age bin']=pd.cut(df['Age'],bins=[0,12,18,50,100],labels=['Child', 'Teen', 'Adult', 'Senior'])
df['Age bin']
```

```python
df['FareBin'] = pd.qcut(df['Fare'], q=4, labels=['Low','Medium','High','Very High'])
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['Fare'] = np.where(df['Fare'] > upper, upper,
                      np.where(df['Fare'] < lower, lower, df['Fare']))
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
df[['Age', 'Fare','Family size']] = scaler.fit_transform(df[['Age','Fare','Family size']])
```

```python
df = pd.get_dummies(df, columns=['Sex','Embarked','Age bin','FareBin'], drop_first=True)
```
